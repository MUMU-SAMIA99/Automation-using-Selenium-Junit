# Dlib (Detecting faces with HOG)

```python
import dlib
import cv2
from google.colab.patches import cv2_imshow
```

```python
pic_1=cv2.imread('/content/drive/MyDrive/Project python for Govt/Practice Document/Team.webp')
cv2_imshow(pic_1)
pic_1.shape
```

```python
face_detector_hog=dlib.get_frontal_face_detector()
detect=face_detector_hog(pic_1,1)
```

```python
detect
```

```python
len(detect)
```

```python
for face in detect:
   l,t,r,b=face.left(),face.top(),face.right(),face.bottom()
   cv2.rectangle(pic_1,(l,t),(r,b),(0,255,255),2)
cv2_imshow(pic_1)
```
